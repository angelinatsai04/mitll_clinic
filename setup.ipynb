```python
# ============================================
# STEP 1: INITIAL SETUP (Run once per new pod)
# ============================================

import sys
import os
import subprocess

# Configuration
INSTALL_DIR = '/workspace/python_packages'
CACHE_DIR = '/workspace/.cache/pip'
TMP_DIR = '/workspace/tmp'

# Create directories
for dir_path in [INSTALL_DIR, CACHE_DIR, TMP_DIR]:
    os.makedirs(dir_path, exist_ok=True)
    print(f"✓ Created: {dir_path}")

# Set environment variables for this session
os.environ['TMPDIR'] = TMP_DIR
os.environ['PIP_CACHE_DIR'] = CACHE_DIR

# Add to Python path
if INSTALL_DIR not in sys.path:
    sys.path.insert(0, INSTALL_DIR)

print(f"\n📦 Installing packages to: {INSTALL_DIR}")
print("This may take 5-10 minutes...\n")

# Define packages to install
packages = [
    "torch==2.3.1 torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118",
    "transformers==4.44.2",
    "datasets",
    "scikit-learn",
    "scipy",
    "matplotlib",
    "seaborn",
    "persim",
    "ripser"
]

# Install each package
for i, package in enumerate(packages, 1):
    print(f"\n[{i}/{len(packages)}] Installing {package.split()[0]}...")
    
    cmd = [
        sys.executable, "-m", "pip", "install",
        "--target", INSTALL_DIR,
        "--cache-dir", CACHE_DIR,
        "--no-warn-script-location"
    ]
    
    # Add package specification
    cmd.extend(package.split())
    
    try:
        result = subprocess.run(
            cmd,
            capture_output=True,
            text=True,
            timeout=600  # 10 minute timeout per package
        )
        
        if result.returncode == 0:
            print(f"  ✓ {package.split()[0]} installed successfully")
        else:
            print(f"  ✗ Error installing {package.split()[0]}")
            print(f"  Error: {result.stderr[:500]}")
    
    except subprocess.TimeoutExpired:
        print(f"  ✗ Installation timed out for {package.split()[0]}")
    except Exception as e:
        print(f"  ✗ Exception: {str(e)[:200]}")

print("\n" + "="*60)
print("Installation complete!")
print("="*60)
print(f"\nPackages installed to: {INSTALL_DIR}")
print(f"Python path includes: {INSTALL_DIR in sys.path}")

# Verify installation
print("\nVerifying installations...")
test_imports = [
    "torch",
    "transformers",
    "datasets",
    "sklearn",
    "scipy",
    "matplotlib",
    "seaborn",
    "persim",
    "ripser"
]

success_count = 0
for module in test_imports:
    try:
        __import__(module)
        print(f"  ✓ {module}")
        success_count += 1
    except ImportError:
        print(f"  ✗ {module} - FAILED")

print(f"\n{success_count}/{len(test_imports)} packages verified")

if success_count == len(test_imports):
    print("\n🎉 All packages installed and verified successfully!")
else:
    print("\n⚠️  Some packages failed - see errors above")
```
