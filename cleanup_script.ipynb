```python
import shutil
import os

# Define the path to the folder you want to delete
folder_to_delete = "/workspace/experiments/2025-10-31_00-03-18_bert-base-uncased_adversarial_analysis_v1"

# Check if the folder exists before attempting deletion
if os.path.isdir(folder_to_delete):
    try:
        shutil.rmtree(folder_to_delete)
        print(f"Folder '{folder_to_delete}' and its contents deleted successfully.")
    except OSError as e:
        print(f"Error: {e.filename} - {e.strerror}.")
else:
    print(f"Folder '{folder_to_delete}' does not exist.")
```
